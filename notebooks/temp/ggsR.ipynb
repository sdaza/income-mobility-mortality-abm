# GSS income, education, age

```r
library(haven)
library(sdazar)
library(ggplot2)
```

```r
getwd()
```

```r
dat = read_dta('../data/GSS2016.DTA')
```

```r
dat = data.table(dat)
```

```r
hist(log(dat$age))
```

```r
ggplot(dat, aes(x=log(realinc),y=log(educ))) + 
    geom_point()
```

```r
summary(lm(log(age) ~ I(log(realinc)) +  I(log(realinc)^2), data = dat))
```

```r
m = lm(educ ~ I(log(realinc))  + age, data = dat)
summary(m)
```
